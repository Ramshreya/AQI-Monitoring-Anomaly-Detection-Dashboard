```python
import requests
import pandas as pd
import os

file_path = "aqi_live_base.csv"

url = "https://api.waqi.info/feed/hyderabad/?token=demo"
response = requests.get(url)
data = response.json()

new_data = {
    "city": "Hyderabad",
    "aqi": data["data"]["aqi"],
    "datetime": data["data"]["time"]["s"]
}

if os.path.exists(file_path):
    df = pd.read_csv(file_path)
    df = pd.concat([df, pd.DataFrame([new_data])], ignore_index=True)
else:
    df = pd.DataFrame([new_data])

df.to_csv(file_path, index=False)
print("✅ Total rows now:", len(df))
df.tail()
```

```python
import pandas as pd
from datetime import timedelta
import random


df_live = pd.read_csv("aqi_live_base.csv")


base_time = pd.to_datetime(df_live["datetime"].iloc[-1])
base_aqi = int(df_live["aqi"].iloc[-1])

print("Base time:", base_time)
print("Base AQI :", base_aqi)


rows = []
current_time = base_time - timedelta(days=90)   # start 90 days before last reading

while current_time <= base_time:
    
    aqi = base_aqi + random.randint(-40, 40)
    aqi = max(30, min(aqi, 300))  # keep within reasonable range

    rows.append({
        "city": "Hyderabad",
        "datetime": current_time,
        "aqi": aqi
    })

    current_time += timedelta(hours=1)

df_3months = pd.DataFrame(rows)


for i in range(0, len(df_3months), 300):
    df_3months.loc[i, "aqi"] = random.randint(350, 450)


df_3months.to_csv("hyderabad_aqi_3months.csv", index=False)

print("✅ 3-month dataset created!")
print("Total rows:", len(df_3months))
df_3months.head()
```

```python
import pandas as pd

# Load the 3-month dataset
df = pd.read_csv("hyderabad_aqi_3months.csv")


df["datetime"] = pd.to_datetime(df["datetime"])
df = df.sort_values("datetime")

# Check for missing values
print("Missing values:\n", df.isna().sum())

# Remove duplicates if any
df = df.drop_duplicates()


print(df.info())
df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(df["datetime"], df["aqi"])
plt.xlabel("Time")
plt.ylabel("AQI")
plt.title("Hyderabad AQI Trend (3 Months)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
from scipy import stats

# Calculate Z-score
df["z_score"] = stats.zscore(df["aqi"])
df["anomaly"] = df["z_score"].apply(lambda x: 1 if abs(x) > 3 else 0)
print("Total anomalies detected:", df["anomaly"].sum())

df[df["anomaly"] == 1].head()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df["datetime"], df["aqi"], label="AQI")

# Highlight anomalies
anomalies = df[df["anomaly"] == 1]
plt.scatter(anomalies["datetime"], anomalies["aqi"], color="red", label="Anomalies")

plt.xlabel("Time")
plt.ylabel("AQI")
plt.title("AQI Trend with Detected Anomalies")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import streamlit as st

st.title("AQI Monitoring & Anomaly Detection Dashboard")

df = pd.read_csv("hyderabad_aqi_3months.csv")
df["datetime"] = pd.to_datetime(df["datetime"])

df["z_score"] = stats.zscore(df["aqi"])
df["anomaly"] = df["z_score"].apply(lambda x: 1 if abs(x) > 3 else 0)

fig, ax = plt.subplots(figsize=(10,5))
ax.plot(df["datetime"], df["aqi"], label="AQI")
anomalies = df[df["anomaly"] == 1]
ax.scatter(anomalies["datetime"], anomalies["aqi"], color="red", label="Anomalies")

ax.set_xlabel("Time")
ax.set_ylabel("AQI")
ax.legend()
plt.xticks(rotation=45)

st.pyplot(fig)
```
